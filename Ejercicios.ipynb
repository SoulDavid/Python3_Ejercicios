Ejercicio 4: Crea un programa que lea una palabra, la guarde en una lista y compruebe si se trata de un palíndromo.

```python
Palabra_04_T5 = input("Introduzca una palabra: ")
lista_Palabra_04_T5 = []
lista_revertida_04_T5 = []

esPalindromo = True

for character_05_T5 in Palabra_04_T5:
  lista_Palabra_04_T5.append(character_05_T5)

for c in lista_Palabra_04_T5:
  lista_revertida_04_T5.append(c)

lista_Palabra_04_T5.reverse()

print(lista_revertida_04_T5)
print(lista_Palabra_04_T5)

for CHARACTER_05_T5 in range(len(Palabra_04_T5)):
  if lista_Palabra_04_T5[CHARACTER_05_T5] != lista_revertida_04_T5[CHARACTER_05_T5]:
    esPalindromo = False

print("Es palindromo") if esPalindromo else print("No es palindromo")
```

```python
def diccionarios_length(lista_palabras):
  """
  Esta función crea un diccionario que contiene cuántas palabras son de longitud par y cuántas de longitud impar.

  Args:
    lista_palabras(list): lista de palabras
  
  Returns:
    dicc_length: diccionario de longitudes
  """
  dicc_length = {}
  contador_pares = 0
  contador_impares = 0
  for palabra in lista_palabras:
    if len(palabra) % 2 == 0:
      contador_pares+=1
    else:
      contador_impares+=1

  dicc_length["Pares"] = contador_pares
  dicc_length["Impares"] = contador_impares

  return dicc_length
```

```python
def caracteres_string(stringe):
  """
  Esta función cuenta cuántas veces sale cada caracter en dicho string

  Args:
    stringe (string): la palabra/frase
  
  Return:
    letras_usadas(diccionario): los caracteres usadas, y las veces que aparece
  """
  letras_usadas = {}
  for caracter in stringe:
    if caracter == " ":
      letras_usadas[" "] = stringe.count(caracter)
    elif caracter not in letras_usadas.keys():
      letras_usadas[caracter] = stringe.count(caracter)

  return letras_usadas

```
